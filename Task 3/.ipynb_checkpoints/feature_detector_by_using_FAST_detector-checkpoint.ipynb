```python
import cv2
img = cv2.imread("C:/Users/hrida/OneDrive/Desktop/Computer vision/Data set/Shapes/geometry.jpg", 0)
```

```python
# FAST object with default values
detector = cv2.FastFeatureDetector_create(50)   #Detects 50 points
points = detector.detect(img, None)
img2 = cv2.drawKeypoints(img, points, None, flags=0)
```

```python
cv2.imshow('Corners',img2)
cv2.waitKey(0)
```
